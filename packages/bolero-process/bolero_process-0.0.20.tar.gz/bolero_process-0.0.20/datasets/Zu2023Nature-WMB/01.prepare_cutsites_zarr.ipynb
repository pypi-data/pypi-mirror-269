```python
import pathlib
from concurrent.futures import ProcessPoolExecutor, as_completed

from bolero_process.atac.sc.fragments_to_zarr import adata_cutsites_to_zarr
```

```python
adata_paths = {"-".join(p.name.split("_")[:-2]): p for p in pathlib.Path("Zu2023Nature/").glob("*.h5ad")}
```

```python
def run(sample, path):
    output_zarr_path = f"/wmb/zarr/Zu2023Nautre.CutSites/{sample}.zarr"
    success_flag = pathlib.Path(output_zarr_path) / ".success"

    if pathlib.Path(output_zarr_path).exists():
        if success_flag.exists():
            # print(f"Zarr store already exists at {output_zarr_path}")
            return
    adata_cutsites_to_zarr(
        adata_path=str(path),
        output_zarr_path=output_zarr_path,
        barcode_prefix=sample,
        chunk_size=20000000,
    )
    success_flag.touch()


fs = {}
with ProcessPoolExecutor(60) as pool:
    for sample, adata_path in adata_paths.items():
        f = pool.submit(run, sample, adata_path)
        fs[f] = sample
    for i, f in enumerate(as_completed(fs)):
        sample = fs[f]
        print(i, sample)
        f.result()
```
